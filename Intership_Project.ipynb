# Project Title:
#### Exploratory Data Analysis on Drugs, Side Effects, and Medical Conditions

## 1. Objective
 The goal is to analyze the relationships between drugs, their side effects, and the
 medical conditions they treat, as well as to explore the ratings and reviews
 associated with these drugs.
## 2. Dataset Overview
 The dataset contains the following columns:
 
 ● drug_name: Name of the drug.
 
 ● medical_condition: The condition the drug is used to treat.
 
 ● side_effects: Common side effects of the drug.
 
 ● generic_name: The generic name of the drug.
 
 ● drug_classes: The class of the drug (e.g., antibiotic, antihistamine).
 
 ● brand_names: Brand names under which the drug is sold.
 
 ● activity: The activity of the drug (e.g., active, inactive).
 
 ● rx_otc: Indicates if the drug is prescription (Rx) or over-the-counter (OTC).
 
 ● pregnancy_category: The drug's pregnancy risk category.
 
 ● csa: Controlled Substances Act schedule, if applicable.
 
 ● alcohol: Interactions with alcohol.
 
● related_drugs: Other drugs related to the primary drug.

 ● medical_condition_description: A brief description of the medical condition.
 
 ● rating: Average user rating of the drug.
 
 ● no_of_reviews: Number of user reviews.
 
 ● drug_link: URL link to more information about the drug.
 
 ● medical_condition_url: URL link to more information about the medical condition.
 
## 3. Tools Required
 ● Python: The primary programming language for data analysis.
 
 ● Pandas: For data manipulation and analysis.
 
 ● Matplotlib/Seaborn: For data visualization.
 
 ● Jupyter Notebook: To write and run Python code.

# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
data = pd.read_excel('Book1.xlsx')
```

```python
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
data.drop(columns=['drug_link','medical_condition_url'], inplace=True)
```

# Data Cleaning

```python
data.isnull().sum()
```

### Handling missing values
 Drop or fill missing values depending on the context.


```python
data_cleaned=data.dropna()
data_cleaned
```

```python
data_cleaned.isnull().sum()
```

# Basic Data Exploration


```python
data_cleaned.describe()
```

```python
# Distribution of drug ratings
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['rating'], bins=10, kde=True)
plt.title('Distribution of Drug Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

### Analyzing Relationships
  TopDrugs by Condition

```python
# Count the most common drugs for each medical condition
top_drugs =data_cleaned.groupby('medical_condition')['drug_name'].value_counts()
print(top_drugs)
```

```python
# Count the most common severe side effects
severe_side_effects = data_cleaned['side_effects'].value_counts().head(10)

# Plot
plt.figure(figsize=(20, 5))
severe_side_effects.plot(kind='bar', color='red')
plt.xlabel("Severe Side Effects")
plt.ylabel("Count")
plt.title("Top 10 Most Common Severe Side Effects")
plt.xticks(rotation=45)
plt.show()
```

# Correlation Between Drug Rating and Number of Reviews


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data_cleaned['rating'], y=data_cleaned['no_of_reviews'])
plt.xlabel('Drug Rating')
plt.ylabel('Number of Reviews')
plt.title('Correlation Between Drug Rating and Number of Reviews')
plt.show()
```

## Insight:
Popular drugs (higher number of reviews) tend to have a wider range of ratings, as more people have different experiences.

Less popular drugs have fewer reviews, making it difficult to judge their effectiveness based on limited data.

# Pregnancy Category Impact on Drug Rating

```python
pregnancy_rating = data_cleaned.groupby('pregnancy_category')['rating'].mean().sort_values()

print("Average Ratings by Pregnancy Category:")
print(pregnancy_rating)
```

## Insight:
Drugs classified under Category A (Safe) tend to have higher ratings.

Drugs with Category C/D/X (Potential risk to the fetus) may have lower ratings due to concerns about safety.

This helps doctors and pregnant women choose safer medications.

# Prescription (Rx) vs. Over-the-Counter (OTC) Ratings

```python
# Compare the average ratings of prescription vs. OTC drugs
rx_vs_otc = data_cleaned.groupby('rx_otc')['rating'].mean()

print("Average Ratings: Prescription (Rx) vs. OTC Drugs:")
print(rx_vs_otc)
```

## Insight:
OTC drugs tend to have higher ratings as they are generally safe and widely available.

Prescription drugs may have lower ratings due to stronger effects, side effects, or dependency risks.
    
Helps users decide whether they need prescription medication or can opt for OTC alternatives.

# Top 5 Most Prescribed Drug Classes

```python
# Count drug classes
top_drug_classes = data_cleaned['drug_classes'].value_counts().head(5)

print("Top 5 Most Prescribed Drug Classes:")
print(top_drug_classes)
```

## Insight:
The most prescribed drug classes help understand trending medical needs.

If antibiotics or painkillers are among the top, it shows common ailments like infections and chronic pain.
    
Pharmaceutical companies can focus on improving these drug categories.

# Number of Drugs Interacting with Alcohol

```python
# Count drugs that interact with alcohol (assuming 'X' indicates interaction)
alcohol_interactions = data_cleaned[data_cleaned['alcohol'] == 'X'].shape[0]

print(f"Total drugs interacting with alcohol: {alcohol_interactions}")
```

## Insight:
A high number of interactions indicate that patients need to be cautious.

Doctors and pharmacists can warn patients about potential side effects when consuming alcohol with these drugs.

# Medical Condition With the Highest Number of Different Drug Treatments

```python
# Count unique drugs for each medical condition
top_conditions = data_cleaned.groupby('medical_condition')['drug_name'].nunique().sort_values(ascending=False).head(5)

print("Medical Conditions with the Most Treatment Options:")
print(top_conditions)
```

## Insight:
Conditions with multiple treatment options indicate ongoing medical research.
    
If a condition has few available drugs, it suggests the need for further research and drug development.

# What are the top 10 most frequently occurring medical conditions

```python
# Count occurrences of each medical condition
top_conditions = data_cleaned['medical_condition'].value_counts().head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_conditions.values, y=top_conditions.index, palette="viridis")
plt.xlabel("Number of Drugs Associated")
plt.ylabel("Medical Condition")
plt.title("Top 10 Most Common Medical Conditions in the Dataset")
plt.show()
```

#  Insights:
This will highlight the most frequently treated medical conditions in the dataset.

It can help identify which conditions have the most drugs available.

If a condition has significantly more drugs, it may indicate high demand or prevalence

# How do average drug ratings vary across different medical conditions

```python
# Count occurrences of each medical condition (Top 10)
top_conditions = data_cleaned['medical_condition'].value_counts().head(10)

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(top_conditions, labels=top_conditions.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribution of Drugs Across Top 10 Medical Conditions")
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```

# Insights:
This shows which medical conditions have the most drugs associated with them.

Larger slices indicate conditions that are more commonly treated in the dataset.

If one condition dominates, it may be more prevalent or require more drug variations.

# Ratings by Pregnancy Category

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='pregnancy_category', y='rating', data=data_cleaned, palette="coolwarm")
plt.xlabel("Pregnancy Category")
plt.ylabel("Drug Rating")
plt.title("Ratings of Drugs by Pregnancy Category")
plt.show()
```

#  Insights:
Category D & X drugs (which pose risks to pregnancy) tend to have lower ratings.

Safer drugs (Categories A & B) have higher ratings, showing patient preference for safety.

# Most Common Medical Conditions Treated

```python
top_conditions = data_cleaned['medical_condition'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_conditions.index, y=top_conditions.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel("Medical Condition")
plt.ylabel("Number of Drugs")
plt.title("Top 10 Most Common Medical Conditions Treated")
plt.show()
```

##  Insights:
Some medical conditions (like depression, anxiety, and diabetes) dominate the dataset.
    
Other conditions have fewer available drugs, suggesting potential gaps in drug development.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='alcohol', y='rating', data=data_cleaned)
plt.xlabel('Alcohol Interaction')
plt.ylabel('Rating')
plt.title('Effect of Alcohol on Drug Ratings')
plt.show()
```

# Insights
#### 1.Wider Distribution for Alcohol-Interacting Drugs

Drugs that interact with alcohol have a wider spread of ratings, meaning patient experiences vary significantly.

Some users may have had negative side effects, leading to low ratings, while others may have positive experiences, resulting in higher ratings.

#### 2️.Possible Dip in Median Ratings

If the median rating for alcohol-interacting drugs is lower than non-interacting drugs, it suggests patients might dislike restrictions or experience stronger side effects when alcohol is involved.
Users might prefer drugs that allow alcohol consumption without significant health concerns.

#### 3️.Tighter Distribution for Non-Alcohol Interacting Drugs

Drugs that do not interact with alcohol may have a more consistent rating, meaning patients generally have stable experiences with them.
